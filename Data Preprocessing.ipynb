```python
# One Hot Encoder concept 
data_out['Department']=data_out['Department'].astype('category')
data_out['Department'] = data_out['Department'].cat.codes
data_out = pd.get_dummies(data_out, columns = ['Department'])
data_out
```

```python
data_out = data_out.iloc[:,0:4]
data_out
```

```python
Q1 = data_out.quantile(0.25)
Q3 = data_out.quantile(0.75)
IQR = Q3-Q1
```

```python
Q1
```

```python
Q3
```

```python
IQR
```

```python
# outlier 
data_out= data_out[~((data_out<(Q1-1.5*IQR))| (data_out>(Q3+1.5*IQR))).any(axis=1)]
```

```python
sns.boxplot(y='Age', data=data_out)
```
